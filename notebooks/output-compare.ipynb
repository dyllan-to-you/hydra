```python
import pandas as pd
from pathlib import Path
import json
pd.set_option('display.max_columns', None)
pd.reset_option('display.max_columns')
pd.set_option('display.max_colwidth', 0)

pd.set_option('display.max_rows', None)

outputDir = Path('../output')

def parse(output):
    split = output.split()
    file1 = outputDir / split[1]
    file2 = outputDir / split[3]
    return file1, file2

def strip_dupes(df, index=True):
    if index is True:
        index = ['figname']
    df = df.reset_index().set_index(index)
    return df[~df.index.duplicated(keep="first")]

def ld(file, dedup=False, sort=False, save=False):
    df = pd.read_parquet(file)
    # print(df)
    if dedup:
        df = strip_dupes(df, dedup)
    if sort:
        df = df.sort_index()
    if save:
        df.to_parquet(file)
    return df

def load(file1, file2, dedup=False, sort=False, save=False):
    return ld(file1, dedup, sort, save), ld(file2, dedup, sort, save)

def compare(df1, df2, reset_index=False):
    # if len(df1) > len(df2):
    #     return df1.loc[df1.index.difference(df2.index)]
    # elif len(df1) < len(df2):
    #     return df2.loc[df2.index.difference(df1.index)]

    if reset_index:
        df1 = df1.reset_index()
        df2 = df2.reset_index()
    return df1.compare(df2)
```
